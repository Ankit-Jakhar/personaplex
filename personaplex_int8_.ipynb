```python
import sys
sys.path.insert(0, '/content/src/moshi-personaplex/moshi')

import torch
from moshi.models import loaders

model_path = "/root/.cache/huggingface/hub/models--nvidia--personaplex-7b-v1/snapshots/31c4ae40e3710304737e6f7b70df23bd06917e5c"

print("Loading moshi...")
lm = loaders.get_moshi_lm(
    f"{model_path}/model.safetensors",
    device='cuda'
)
print("Loaded!")

print("Quantizing to 8-bit...")
lm = torch.quantization.quantize_dynamic(
    lm, {torch.nn.Linear}, dtype=torch.qint8
)
print("Quantized!")

print("Saving...")
torch.save(lm.state_dict(), "/content/personaplex_quantized.pt")
print("Done!")

import os
size = os.path.getsize("/content/personaplex_quantized.pt") / (1024**3)
print(f"File size: {size:.2f} GB")
```
